```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from scipy import integrate

#%mayplotlib inline
mean = 100
variance = np.sqrt(50)

x = np.linspace(mean - 3 * variance, mean + 3 * variance, 200)
y = np.exp((-(x - mean) ** 2) / (2 * (variance ** 2))) / (np.sqrt(2 * np.pi) * variance)

plt.plot(x, y, "r-", linewidth = 2)
plt.grid(True)
plt.savefig("f3.png",dpi=500)

def f(x):
    return np.exp((-(x - mean) ** 2) / (2 * (variance ** 2))) / (np.sqrt(2 * np.pi) * variance)
print(integrate.quad(f,float('-inf'),float('inf')))
```
