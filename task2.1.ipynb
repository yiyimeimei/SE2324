```python
#from scipy import linalg
import numpy as np
import array as arr
import scipy as sci
import numpy.linalg as lg
import matplotlib.pyplot as plt
def f(x):
    return 1 / (1 + x ** 2)
def Fourier(x, j, n):
    if j >= 1 and j <= (n / 2):
        return np.sin(j * np.pi * x)
    elif j >= ((n / 2) + 1) and j <= n:
        return np.cos((j - n / 2) * np.pi * x)    
    
#N = M = 8
N = 8
m = N - 1 

arrayX = arr.array('d') 
arrayf = arr.array('d')
for i in range (0, N): 
    arrayX.append(i / m)
for i in range (0, N):
    arrayf.append(f(arrayX[i]))

#Vandermonde
V = np.vander(arrayX, increasing = True)
p, l, u = sci.linalg.lu(V)
y = lg.solve(p.dot(l), arrayf)
c1 = lg.solve(u, y)
print("When N = 8, c in Vc = f is \n", c1)
L2norm1 = lg.norm((np.matmul(V, c1) - arrayf))
print("Its L2 norm is ", L2norm1)


#Fourier
F = np.zeros((N, N))
for i in range(0, N):
    for j in range(0, N):
        F[i, j] = Fourier(arrayX[i], j + 1, N)
p, l, u = sci.linalg.lu(F)
y = lg.solve(p.dot(l), arrayf)
c2 = lg.solve(u, y)
print("When N = 8, c in Fc = f is \n", c2)
L2norm2 = lg.norm(np.matmul(F, c2) - arrayf)
print("Its L2 norm is ", L2norm2)


#N = 16
N = 16
m = N - 1 

arrayX = arr.array('d') 
arrayf = arr.array('d')
for i in range (0, N): 
    arrayX.append(i / m)
for i in range (0, N):
    arrayf.append(f(arrayX[i]))
    
    
#Vandermonde
V = np.vander(arrayX, increasing = True)
p, l, u = sci.linalg.lu(V)
y = lg.solve(p.dot(l), arrayf)
c1 = lg.solve(u, y)
print("When N = 16, c in Vc = f is \n", c1)
L2norm1 = lg.norm((np.matmul(V, c1) - arrayf), ord = 2)
print("Its L2 norm is ", L2norm1)

#Fourier
F = np.zeros((N, N))
for i in range(0, N):
    for j in range(0, N):
        F[i, j] = Fourier(arrayX[i], j + 1, N)
p, l, u = sci.linalg.lu(F)
y = lg.solve(p.dot(l), arrayf)
c2 = lg.solve(u, y)
#c2 = lg.solve(F, arrayf)
print("When N = 16, c in Fc = f is \n", c2)
L2norm2 = lg.norm(np.matmul(F, c2) - arrayf)
print("Its L2 norm is ", L2norm2)
```
